```python
import boto3
from dotenv import load_dotenv, find_dotenv
import os
```

```python
load_dotenv()

API_KEY_S3 = os.environ["API_KEY_S3"]
API_SECRET_KEY_S3 = os.environ["API_SECRET_KEY_S3"]
```

```python
session = boto3.Session(aws_access_key_id = API_KEY_S3 , 
                        aws_secret_access_key = API_SECRET_KEY_S3,
                        region_name = "eu-west-3")
print(session)
```

```python
session = session.resource("s3") 
```

```python
bucket_name = "dsfs36-bucket-01"
file_name = "../data/outputs/booking_hotels_top5_cities.csv"
object_name = "outputs/booking_hotels_top5_cities.csv"

session.Bucket(bucket_name).upload_file(file_name, object_name)
print(f"Fichier {file_name} uploadé dans s3://{bucket_name}/{object_name}")
```
